```python
#import statements
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder,LabelEncoder
import pickle
```

```python
#READ THE DATA
data=pd.read_csv("CM.csv")
data.head()
```

```python
#drop the unnecessary coloumns 
data=data.drop(['RowNumber','CustomerId','Surname'],axis=1) #but defaut it considers axis=0 which means drop by rows
data
```

```python
#convert into 0 and 1
le=LabelEncoder()
data['Gender']=le.fit_transform(data['Gender'])#if you want to perform any thing with colounms please use data['col_name']
data
```

```python
# data['Geography'].shape
# Attribute = data, Method = behavior, Function = general-purpose code(please undersatnd this)
# data[['Geography']].shape
ohe=OneHotEncoder()

#converting it to binary form and adding it to veriable
geoencoder=ohe.fit_transform(data[['Geography']]) #---> should be with double [[]] quotes


#making it into 3 columns by getting the feature out
#coloumns=ohe.get_feature_names_out(['Geography'])

#converting it into array as it is sparse
# geoencoder.toarray()

#need to convert it into dataframe 
geoencoder=pd.DataFrame(geoencoder.toarray(),columns=ohe.get_feature_names_out(['Geography']))

#first drop the geography veriable the join the data with geoencoder
data=data.drop(['Geography'],axis=1)
data=pd.concat([geoencoder,data],axis=1)

```

```python
data
```

```python
# convert labelencoder and onhotencoder as pkl file 
with open('label_encoder.pkl','wb') as file:
    pickle.dump(le,file)
with open('one_hot_encoder.pkl','wb') as file:
    pickle.dump(ohe,file)
```

```python
# dividing the data for train test split
x=data.drop('Exited',axis=1)
y=data['Exited']

x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)

scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)

with open('scaler.pkl','wb') as file:
    pickle.dump(scaler,file)
    
```

ANN IMPLEMENTATION


```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime
```

```python
x_train.shape[1] #this gives you no of columns in x train
```

```python
#how to build a model

# Sequential(...) → round brackets () mean:→ “Call the Sequential constructor with these arguments”
# Inside it, [ ... ] → square brackets [] mean:→ “This is a Python list of layers”

# ___Keras expects input_shape as a tuple, like (num_features,), so for one-dimensional input you must write the comma.___
model=Sequential([
    Dense(64,activation='relu',input_shape=(x_train.shape[1],)), 
    Dense(32,activation='relu'),
    Dense(1,activation='sigmoid')
])
#check the model summery for the features
model.summary()
```

```python
opt=tf.keras.optimizers.Adam(learning_rate=0.01)
model.compile(optimizer=opt,loss="binary_crossentropy",metrics=['accuracy'])
```

```python
# THIS WILL TELL HOW OUR LOGS ARE GOING TO BE
logs_dir="logs/fit"+datetime.datetime.now().strftime("%y%m%d-%H%M%S")

#THIS WILL VISUALIZE THE LOGS HIVEN TO US USING HISTOGRAM
tensor_callback=TensorBoard(log_dir=logs_dir,histogram_freq=1,write_graph=False)

#THIS WE WILL USE TO STOP THE EPOCH IF NO BIG IMPOROVEMENT IS GOING ON AND ALS IT WILL MAKE RESTORE THE BEST WEIGHTS
tensor_earlystopping=EarlyStopping(monitor='val_loss',patience=5,restore_best_weights=True)
```

```python
history=model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=100,
                  callbacks=[tensor_callback,tensor_earlystopping])
```

```python
# SAVE THE MODEL
model.save("model.h5")
```
